##Classroom Attendance Management System


```python
class_hour=(8,9,10,11,12)#tuple
student_list=["Ali","hassan","jawad","shafiq","zeeshan","safdar","hasnain"]#list
Authoriz_staff={"nabeel","zara","zohaib"}#set
Attendence_record={}
print("---------Wlcome to the Attendence Managment system---------")
user=input("Enter name of user to asses the system")
if user in Authoriz_staff:
    print("Assess Granted.",user)
    running=True
    while running:
        print("1. Take Attendence")
        print("2. view student list")
        print("3. view attendence record")
        print("4. Exit program. Goodbye!!!")
        choice=input("Enter your choice")
        if choice=='1':
            print("Mark the attendence(p=present,A=Absent)")
            for student in student_list:
                status=input("Is " +student+ " Present(P/A)").upper()
                if status=="P":
                    Attendence_record[student]="Present"
                else:
                    Attendence_record[student]="Absent"
            print("Attendence record.")
        elif choice=='2':
            print("---Current Student---")
            for idx , student in enumerate(student_list):
                print(idx+1,student)
        elif choice=='3':
            print("Attendence record")
            if not Attendence_record:
                print("No yet record")
            else:
                for student,status in Attendence_record.items():
                    print(student,status)
        elif choice=='4':
            print("Exiting the program.Goodbye!!!")
            running=False
        else:
            print("Invalid chhoice . please try again")
    print("----Summary----")
    print("Class hour today:")
    for hour in class_hour:
        print(hour,":00 AM")
else:
    print("Assess Denied.")
```

## Question about python

```python
score = 0
total_questions = 3

# 2. String Manipulation & Output
print("--- Welcome to the Python Basics Quiz! ---")
user_name = input("Enter your name: ")
print(f"Hello {user_name}, let's start!\n")

# Questions data: list of dictionaries
questions = [
    {"q": "What is the output of print(2**3)?", "a": "8"},
    {"q": "Does Python use {} for blocks? (yes/no)", "a": "no"},
    {"q": "Which data type is True/False? (str/int/bool)", "a": "bool"}
]

# 3. For Loop
for i, quest in enumerate(questions):
    print(f"Question {i+1}: {quest['q']}")

    # 4. Input & Casting
    user_answer = input("Answer: ").lower().strip()

    # 5. Conditions & Operators
    if user_answer == quest['a']:
        print("Correct!\n")
        score += 1 # Arithmetic operator
    else:
        print(f"Wrong! The answer was {quest['a']}\n")

# 6. Final Evaluation
print(f"--- Quiz Over, {user_name}! ---")
print(f"Final Score: {score}/{total_questions}")
# Nested Condition
if score == total_questions:
    print("Perfect Score!")
elif score > 0:
    print("Good job!")
else:
    print("Better luck next time!")
```

## Generate strong pasword

```python
import random
import string
print("------Generate strong password-------")
length = int(input("enter the length of password(e.g ,= '16'):"))
use_letters=input("Include letters")
use_numbers=input("Include numbers")
use_symbol=input("Include symbols")
ali_brand=""
if use_letters:
    ali_brand +=string.ascii_letters
if use_numbers:
    ali_brand +=string.digits
if use_symbol:
    ali_brand +=string.punctuation
if not ali_brand:
    print("Error:please select the option one")
    exit()
password="".join(random.choices(ali_brand,k=length))
print("-"*30)
print("Generated password",password)
print("-"*30)
```

##While loop and List project

```python
tasks = ["ali" ,34,656,466,898,89]

while True:
    print("\n-----The Menu-----")
    print("1 To view Task")
    print("2 To Add task")
    print("3 To remove task")
    print("4 To Quit")

    choice = input("Choose the number: ")

    if choice == '1':
        print("Current tasks")
        if not tasks:
            print("There is no task")
        else:
            for i, task in enumerate(tasks, start=1):
                print(i, task)

    elif choice == '2':
        task = input("Enter the new task: ")
        if task:
            tasks.append(task)
            print("Task", task, "added successfully")
        else:
            print("Task cannot be empty")

    elif choice == '3':
        if not tasks:
            print("No task to delete")
        else:
            try:
                task_num = int(input("Enter the number of the task to delete: "))
                if 1 <= task_num <= len(tasks):
                    deleted = tasks.pop(task_num - 1)
                    print("Deleted task:", deleted)
                else:
                    print("Invalid task number")
            except ValueError:
                print("Please enter a valid number")

    elif choice == '4':
        print("Exit the program! Goodbye!")
        break

    else:
        print("Invalid choice, enter a number between 1-4")
```
